# Cifar 10

```python
import sys
sys.path.append('..')

import torch
import matplotlib.pyplot as plt
from Utils.cifar10_loaders import get_cifar10_dataloaders

train_loader, test_loader = get_cifar10_dataloaders()

train_dataset = train_loader.dataset
test_dataset = test_loader.dataset

classes = train_dataset.dataset.classes if hasattr(train_dataset, 'dataset') else train_dataset.classes
num_classes = len(classes)

print("Dataset: CIFAR-10")
print(f"Number of classes: {num_classes}")
print(f"Class names: {classes}")
print(f"Train set size: {len(train_dataset)}")
print(f"Test set size: {len(test_dataset)}")

images, labels = next(iter(train_loader))

def imshow(img_tensor, title):
    img = img_tensor.permute(1, 2, 0).numpy()
    plt.imshow((img * 0.5) + 0.5)
    plt.title(title)
    plt.axis('off')

plt.figure(figsize=(15, 6))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    imshow(images[i], classes[labels[i]])
plt.tight_layout()
plt.show()

print(f"Image shape (C, H, W): {images[0].shape}")
```

# Mnist

```python
import sys
sys.path.append('..')

import torch
import matplotlib.pyplot as plt
from Utils.mnist_loaders import get_mnist_dataloaders
import torchvision.transforms as transforms
from collections import Counter

custom_transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.Grayscale(num_output_channels=1),
    transforms.ToTensor(),
])

train_loader, test_loader = get_mnist_dataloaders(transform_train=custom_transform, transform_test=custom_transform)

train_dataset = train_loader.dataset
test_dataset = test_loader.dataset

label_counts = Counter(label for _, label in train_dataset)
num_classes = len(label_counts)
classes = sorted(label_counts.keys())

print("Dataset: MNIST")
print(f"Number of classes: {num_classes}")
print(f"Class labels: {classes}")
print(f"Train set size: {len(train_dataset)}")
print(f"Test set size: {len(test_dataset)}")

images, labels = next(iter(train_loader))

def imshow(img_tensor, title):
    img = img_tensor[0].numpy()
    plt.imshow((img * 0.5) + 0.5, cmap='gray')
    plt.title(title, fontsize=10)
    plt.axis('off')

plt.figure(figsize=(15, 6))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    imshow(images[i], str(labels[i].item()))
plt.tight_layout()
plt.show()

print(f"Image shape (C, H, W): {images[0].shape}")
```

# Flowers 102

```python
import sys
sys.path.append('..')

import torch
import matplotlib.pyplot as plt
from Utils.flowers102_loaders import get_flowers102_dataloaders
import torchvision.transforms as transforms
from collections import Counter

custom_transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
])

train_loader, test_loader = get_flowers102_dataloaders(
    transform_train=custom_transform,
    transform_test=custom_transform
)

train_dataset = train_loader.dataset
test_dataset = test_loader.dataset

label_counts = Counter(label for _, label in train_dataset)
num_classes = len(label_counts)
classes = sorted(label_counts.keys())

print("Dataset: Flowers102")
print(f"Number of classes: {num_classes}")
print(f"Label indices: {classes[:10]}...")
print(f"Train set size: {len(train_dataset)}")
print(f"Test set size: {len(test_dataset)}")

images, labels = next(iter(train_loader))

def imshow(img_tensor, title):
    img = img_tensor.permute(1, 2, 0).numpy()
    # img = (img * [0.229, 0.224, 0.225]) + [0.485, 0.456, 0.406]  
    img = img.clip(0, 1)
    plt.imshow(img)
    plt.title(f"Class {title}", fontsize=9)
    plt.axis('off')

plt.figure(figsize=(15, 6))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    imshow(images[i], labels[i].item())
plt.tight_layout()
plt.show()

print(f"Image shape (C, H, W): {images[0].shape}")
```
